```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
plt.style.use('classic')
```

```
rs = np.random.RandomState(0)
x = np.linspace(0, 10, 500)
y = np.cumsum(rs.randn(500, 6), 0)
plt.plot(x, y)
plt.legend('ABCDEF', ncol=2, loc='upper');
```
